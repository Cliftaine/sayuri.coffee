## Part B

### Loading files and exploring data

```python
FILE_B= 'Parte B/Examen Parte B.csv'
b_df = pd.read_csv(FILE_B)
print(b_df.shape)
print(b_df.dtypes)
b_df.head(10)
```

### b) Casting Data

```python
b_df['Mes Operacion'] = pd.to_datetime(b_df['Mes Operacion'])
b_df['# Transacciones'] = pd.to_numeric(b_df['# Transacciones']
                                        .str.replace(',',''), downcast='integer')

b_df['Monto Transaccion'] = pd.to_numeric(b_df['Monto Transaccion']
                                            .str.replace(',','')
                                            .str.replace('$','')
                                            , downcast='float')

b_df['Monto Contracargo (Venta)'] = pd.to_numeric(b_df['Monto Contracargo (Venta)']
                                                    .str.replace(',','')
                                                    .str.replace('$','')
                                                    , downcast='float')
```

```python
print(b_df.dtypes)
b_df.head(10)
```

### b) Filtering by MXN and USD

```python
b_df = b_df[b_df['Moneda'].isin(['MXN', 'USD'])]
b_df_mxn = b_df[b_df['Moneda'] == 'MXN']
b_df_usd = b_df[b_df['Moneda'] == 'USD']

print(b_df['Moneda'].unique())
```

### c) USD -> MXN

```python
TOKEN = '13eea40e3b67b03196b6ad3aab04fac4313d82710d611460c07fb9dbbcc197b7'
URL = f'https://www.banxico.org.mx/SieAPIRest/service/v1/series/SF43718/datos'

r=requests.get(URL, headers={"Bmx-Token":TOKEN})
data = r.json()
mxn_usd_df = pd.DataFrame.from_dict(data['bmx']['series'][0]['datos'])
mxn_usd_df['fecha'] = pd.to_datetime(mxn_usd_df['fecha'])
mxn_usd_df['dato'] = pd.to_numeric(mxn_usd_df['dato'] , downcast='float')
grouped = mxn_usd_df.assign(grouper=mxn_usd_df['fecha'].dt.to_period('M'))
grouped = grouped.groupby(grouped['grouper'])['dato'].mean()
grouped.head()

b_df_usd = pd.merge(b_df_usd.assign(grouper=b_df_usd['Mes Operacion'].dt.to_period('M')),
                grouped,
                how='left', on='grouper')

b_df_usd['Monto Transaccion'] = b_df_usd['Monto Transaccion'] * b_df_usd['dato']
b_df_usd['Monto Contracargo (Venta)'] = b_df_usd['Monto Contracargo (Venta)'] * b_df_usd['dato']
b_df_usd = b_df_usd.drop('grouper', 1)
b_df_usd = b_df_usd.drop('dato', 1)
b_df_usd['Moneda'] = 'MXN'
b_df_usd.head()
```

```python
b_df = pd.concat([b_df_mxn, b_df_usd])
```
